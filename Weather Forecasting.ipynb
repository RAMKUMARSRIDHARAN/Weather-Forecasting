## Unique columns for Each columns in weather Dataset

```python
# Iterate through each column
for column in Weather_Forecast.columns:
    # Get unique values and their counts for the column
    unique_values = Weather_Forecast[column].value_counts()
    
    print(f"\nUnique values for column '{column}':")
    print("-" * 100)
    print(unique_values)
```

## Duplicates Detecting for each columns 

```python
# Iterate through each column
for column in Weather_Forecast.columns:
    # Get unique values and their counts for the column
    duplicated_columns = Weather_Forecast[column].duplicated()
    
    print(f"\duplicated values for column '{column}':")
    print("-" * 100)
    print(duplicated_columns.sum())
```

## Minimum and maximum for each columns with for loop

```python
# Iterate through each column
for column in Weather_Forecast.columns:
    # Check if the column is numeric
    if pd.api.types.is_numeric_dtype(Weather_Forecast[column]):
        min_value = Weather_Forecast[column].min()
        max_value = Weather_Forecast[column].max()
        print(f"\nMinimum and Maximum values for column '{column}':")
        print("-" * 100)
        print(f"Min: {min_value}")
        print(f"Max: {max_value}")
    else:
        print(f"\nColumn '{column}' is not numeric, skipping min/max calculation.")
```

# Time Series Algorithm

```python
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Exploratory Data Analysis (EDA)
plt.figure(figsize=(12, 6))
plt.plot(Weather_Forecast['Precip Type'])  # Replace with relevant column name
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.show()
```

```python
# Check for Stationarity
result = adfuller(Weather_Forecast['Temperature (C)'])  # Use specific column for analysis
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

```python
# Decomposition
decomposition = seasonal_decompose(Weather_Forecast['Temperature (C)'], model='additive', period=365)  # Specify period if necessary
decomposition.plot()
plt.show()
```

```python
# Autocorrelation and Partial Autocorrelation Plots
plot_acf(Weather_Forecast['Temperature (C)'])
plot_pacf(Weather_Forecast['Temperature (C)'])
plt.show()
```

```python
# Model Building and Forecasting
model = ARIMA(Weather_Forecast['Temperature (C)'], order=(5,1,0))  # Example ARIMA model parameters
model_fit = model.fit()
print(model_fit.summary())
```

```python
# Visualize the forecast (Note: Forecasting will require predicting future values)
forecast = model_fit.forecast(steps=10)  # Forecast next 10 steps
plt.figure(figsize=(12, 6))
plt.plot(Weather_Forecast['Temperature (C)'], label='Observed')
plt.plot(range(len(Weather_Forecast), len(Weather_Forecast) + 10), forecast, label='Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.title('Temperature Forecast')
plt.show()
```
